# Cleaning Face Covering Data


```python
import pandas as pd
# Import dataset
dataset_url = "https://raw.githubusercontent.com/KrisChuY/Data-of-Application/main/Policy_face-covering-policies-covid.csv"
face_covering = pd.read_csv(dataset_url)
face_covering.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# if uising a Jupyter notebook, include:
%matplotlib inline
```

## Face Covering Step 1: Format Date and Extract Year

```python
# convert date format to datetime
face_covering['Day'] = pd.to_datetime(face_covering['Day'], format='%Y/%m/%d')

# Extract year from date column
face_covering['year'] = face_covering['Day'].dt.year

# check that it works
face_covering.head()
```

## Face Covering Step 2: Transforming the Data, Aggregating Mode, and Compatible Column Names

```python
# create new df with the avg of the face covering index by country and year
face_covering_mode = face_covering.groupby(['Entity', 'year'])['facial_coverings'].agg(pd.Series.mode)

# reset the index to turn the index into columns
face_covering_agg = face_covering_mode.reset_index()

# rename the 'Entity' column to 'Country name' to prep for joining
face_covering_agg.rename(columns={'Entity': 'Country name'}, inplace=True)

# check df to see if it works
face_covering_agg
```

```python
face_covering_agg.to_csv('cleaned_face_covering_final.csv', index=False)
```

# Cleaning Happiness Data

```python
#import libraries
import csv
import pandas as pd
import numpy as np
```

```python
#Import dataset
dataset_url = "https://raw.githubusercontent.com/KrisChuY/Data-of-Application/main/DataForTable2.1WHR2023.xlsx%20-%20Sheet1.csv"
df = pd.read_csv(dataset_url)
df.head()
```

## Happiness Step 1: Extract Years, Group by Year and Country, and Drop Countries with Missing Values

```python
#Filter the dataset that only contains countries with 2020, 2021 and 2022 data
df_filtered = df[df['year'].isin([2020, 2021, 2022])]
grouped = df_filtered.groupby('Country name')['year'].nunique()
selected_countries = grouped[grouped == 3].index.tolist()
df_final = df_filtered[df_filtered['Country name'].isin(selected_countries)]
```

```python
#check the types of the variable
df_final.dtypes
```

```python
#checking missing value

df_final.isnull().sum()
```

```python
#replace missing value with mean
df_final.loc[:, 'Log GDP per capita'] = df_final['Log GDP per capita'].fillna(df_final['Log GDP per capita'].mode()[0])
df_final.loc[:, 'Social support'] = df_final['Social support'].fillna(df_final['Social support'].mode()[0])
df_final.loc[:, 'Healthy life expectancy at birth'] = df_final['Healthy life expectancy at birth'].fillna(df_final['Healthy life expectancy at birth'].mode()[0])
df_final.loc[:, 'Freedom to make life choices'] = df_final['Freedom to make life choices'].fillna(df_final['Freedom to make life choices'].mode()[0])
df_final.loc[:, 'Generosity'] = df_final['Generosity'].fillna(df_final['Generosity'].mode()[0])
df_final.loc[:, 'Perceptions of corruption'] = df_final['Perceptions of corruption'].fillna(df_final['Perceptions of corruption'].mode()[0])
df_final.loc[:, 'Positive affect'] = df_final['Positive affect'].fillna(df_final['Positive affect'].mode()[0])
df_final.loc[:, 'Negative affect'] = df_final['Negative affect'].fillna(df_final['Negative affect'].mode()[0])
```

```python
#check again if the replacement works 
df_final.isnull().sum()
```

```python
#Reset the index after filtering
df_final = df_final.reset_index(drop=True)
df_final.index += 1


df_final
```

```python
#To identify how many countries left after the filtering
num_countries = df_final['Country name'].unique()

num_elements = len(num_countries)

print(num_elements)
```

## Step 2: Deeper Data Cleaning and Removing Unnecessary Columns

```python
#Removing Unnecessary Variables
new_df = df_final.drop(['Positive affect', 'Negative affect'], axis=1)

new_df
```

```python
new_df.to_csv('Cleaned_Happiness.csv', index=False)
```

```python
#import libraries
import csv
import pandas as pd
import numpy as np
```

```python
#Import final dataset
dataset_url = "https://raw.githubusercontent.com/KrisChuY/Data-of-Application/main/merged_df_test.csv"
df = pd.read_csv(dataset_url)
df.head()
```

# Overall Correlation Coefficient of Stringency index and Life Ladder

```python
# Group the data by year
grouped_data = df.groupby('year')

# Concatenate all data from different years into a single DataFrame
all_data = pd.concat([data for _, data in grouped_data])

# Calculate the correlation coefficient for the entire dataset
correlation_collective = all_data['stringency_index'].corr(all_data['Life Ladder'])

# Print the result for the collective analysis
print(f"Collective Analysis - Correlation: {correlation_collective}")
```

# Correlation Coefficient of Stringency index and Life Ladder Per year

```python
# Iterate over each group
for year, data in grouped_data:
    # Select the 'stringency_index' and 'Life Ladder' columns for the current year
    year_data = data[['stringency_index', 'Life Ladder']]
    
    # Calculate the correlation coefficient
    correlation = year_data.corr().iloc[0, 1]  # Get the correlation value
    
    # Print the result for the current year
    print(f"Correlation for year {year}: {correlation}")
```

#Overall Correlation Coefficient bewteen Restriction Policies and Life Ladder

```python
# Concatenate all data from different years into a single DataFrame
all_data = pd.concat([data for _, data in grouped_data])

# Calculate the correlation coefficient for the entire dataset (facial_coverings)
correlation_collective_coverings = all_data['facial_coverings'].corr(all_data['Life Ladder'])

# Calculate the correlation coefficient for the entire dataset (cancel_public_events)
correlation_collective_cancel_events = all_data['cancel_public_events'].corr(all_data['Life Ladder'])

# Calculate the correlation coefficient for the entire dataset (restriction_gatherings)
correlation_collective_restrict_gatherings = all_data['restriction_gatherings'].corr(all_data['Life Ladder'])

# Print the results for the collective analysis
print(f"Collective Analysis - facial_coverings Correlation: {correlation_collective_coverings}")
print(f"Collective Analysis - cancel_public_events Correlation: {correlation_collective_cancel_events}")
print(f"Collective Analysis - restriction_gatherings Correlation: {correlation_collective_restrict_gatherings}")
```

#Overall Correlation Coefficient bewteen Restriction Policies and Life Ladder per year

```python
# Create an empty DataFrame to store the correlation results
correlation_results = pd.DataFrame(columns=['Policy', 'Year', 'Correlation'])

# Iterate over each group
for year, data in grouped_data:
    # Iterate over each policy variable
    for policy in ['facial_coverings', 'cancel_public_events', 'restriction_gatherings']:
        # Select the policy variable and 'Life Ladder' columns for the current year
        year_data = data[[policy, 'Life Ladder']]
        
        # Calculate the correlation coefficient
        correlation = year_data.corr().iloc[0, 1]  # Get the correlation value
        
        # Create a new DataFrame with the correlation result
        result_df = pd.DataFrame({
            'Policy': [policy],
            'Year': [year],
            'Correlation': [correlation]
        })
        
        # Concatenate the result_df with correlation_results DataFrame
        correlation_results = pd.concat([correlation_results, result_df], ignore_index=True)

# Sort the correlation_results DataFrame by 'Policy' column
correlation_results = correlation_results.sort_values(by='Policy')

# Print the correlation results
for index, row in correlation_results.iterrows():
    print(f"Correlation for {row['Year']}, {row['Policy']}: {row['Correlation']}")
```

# Regression Analysis of Demographic Variables from Happiness Dataset and Life Ladder

```python
import statsmodels.api as sm

# Select the independent variables (X) and the dependent variable (y)
X = df[['Log GDP per capita', 'Social support', 'Healthy life expectancy at birth', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
y = df['Life Ladder']

# Add a constant term to the independent variable
X = sm.add_constant(X)

# Fit the regression model with the interaction terms
model = sm.OLS(y, X).fit()

# Print the regression results
print(model.summary())
```

# Regression Analysis of Stringency Index and Life Ladder

```python
import statsmodels.api as sm

# Select the independent variable (X) and the dependent variable (y)
X = df['stringency_index']
y = df['Life Ladder']

# Select the control variables
control_vars = ['Log GDP per capita', 'Social support', 
                'Freedom to make life choices', 'Perceptions of corruption']

# Add the control variables to the independent variable
X = sm.add_constant(X)
X[control_vars] = df[control_vars]

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression results
print(model.summary())
```
